```python
import os
import pandas as pd
from PIL import Image
import mahotas as mh
import numpy as np


def extract_glcm_features(image_path):
    # Load the image and convert to grayscale
    image = Image.open(image_path).convert('L')
    image_array = np.array(image)

    # Compute Haralick texture features
    haralick_features = mh.features.texture.haralick(image_array, return_mean=True)

    # Extract specific Haralick features
    contrast = haralick_features[1]       # Contrast (Haralick's "contrast")
    dissimilarity = haralick_features[4]   # Dissimilarity
    homogeneity = haralick_features[8]     # Homogeneity (Haralick's "inverse difference moment")
    energy = haralick_features[0]          # Energy (Haralick's "angular second moment")
    correlation = haralick_features[2]     # Correlation

    return [contrast, dissimilarity, homogeneity, energy, correlation]


def process_images(image_dir):
    image_features = []

    # Loop through 'benign' and 'malignant' categories
    for label in ['benign', 'malignant']:
        label_dir = os.path.join(image_dir, label)
        
        # Check if the directory exists
        if not os.path.exists(label_dir):
            print(f"Directory {label_dir} does not exist. Skipping...")
            continue
        
        # Traverse all subdirectories and files
        for root, dirs, files in os.walk(label_dir):
            for file in files:
                # Skip system files with '._' prefix (macOS hidden files)
                if file.startswith('._'):
                    continue

                # Check if file is an image based on its extension
                if file.endswith(('.png', '.jpg', '.jpeg', '.tiff')):
                    image_path = os.path.join(root, file)

                    try:
                        # Extract GLCM features from the image
                        features = extract_glcm_features(image_path)
                        
                        # Append the label and features to the list
                        image_features.append([label] + features)
                    except Exception as e:
                        print(f"Error processing {image_path}: {e}")
                        continue

    # Convert list to DataFrame with appropriate column names
    image_features_df = pd.DataFrame(
        image_features, 
        columns=['label', 'contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation']
    )
    
    return image_features_df


# Set the base directory path for the images
image_dir = '/Volumes/T7/Year 3 Docs/Sem 5/LiveProject/histopathology_images'

# Process images and create a DataFrame with extracted features
image_features_df = process_images(image_dir)

# Show the resulting DataFrame (or save to a CSV if needed)
print(image_features_df.head())
```

```python
image_features_df.tail()
```

```python
# Aggregate image features by label
aggregated_features = image_features_df.groupby('label').mean().reset_index()
print(aggregated_features)
```

```python
# Convert aggregated features to a dictionary for easy mapping
features_dict = aggregated_features.set_index('label').to_dict(orient='index')
features_dict
```

```python
benign_features = features_dict['benign']
malignant_features = features_dict['malignant']
print("benign features: ", benign_features)
print("malignant features: ", malignant_features)
```

```python
# Add benign features to the negative (benign) subset
for feature, value in benign_features.items():
    print(f"feature: {feature} value: {value}")
    clinical_df_negative[feature] = value
```

```python
# Add malignant features to the positive (malignant) subset
for feature, value in malignant_features.items():
    clinical_df_positive[feature] = value
```

```python
# Combine the two subsets into a single DataFrame
combined_clinical_df = pd.concat([clinical_df_positive, clinical_df_negative], ignore_index=True)
combined_clinical_df.drop("id", axis = 1)
```

#### DATA PREPROCESSING

```python
X = combined_clinical_df.copy()

y = X['diagnosis']

X = X.drop(['diagnosis', 'id'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 1234)
```

```python
from imblearn.over_sampling import SMOTE # Synthetic Minority Over-Sampling Techniques

sm = SMOTE(random_state=1234)

X_sm, y_sm = sm.fit_resample(X_train, y_train)

print(f'''Shape of X before SMOTE: {X.shape}
Shape of X after SMOTE: {X_sm.shape}''')

print('\nBalance of positive and negative classes (%):')
y_sm.value_counts(normalize=True) * 100
```

```python
# Import ML Libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score , confusion_matrix, f1_score, roc_auc_score
```

```python
RandomForest_Model = RandomForestClassifier(n_estimators = 100, random_state = 53)
```

```python
RandomForest_Model.fit(X_sm, y_sm)

y_pred = RandomForest_Model.predict(X_test)

print ('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print("Accuracy : ", accuracy_score(y_test, y_pred) *  100)
print("Recall : ", recall_score(y_test, y_pred) *  100)
print("Precision : ", precision_score(y_test, y_pred) *  100)
print("F1 : ", f1_score(y_test, y_pred) *  100)
print("ROC AUC : ", roc_auc_score(y_test, y_pred) *  100)
print('\n\n')
```
